```python
import matplotlib.pyplot as plt

plt.plot([2,4,6,8,11]) # it is default y axis
# x axis automatically created if you dont create .
plt.ylabel("numbers")
plt.xlabel("indoces")
plt.title("simple plot")
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16])
plt.xlabel("Numbers")
plt.ylabel("Square Numbers")
plt.title("simple plot")
plt.grid() # grid format plot.
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16],'ro')
plt.xlabel("Numbers")
plt.ylabel("Square Numbers")
plt.title("simple plot")
plt.grid() # grid format plot.
plt.show()
```

```python
import numpy as np
t = np.arange(0.,5.,0.2)
plt.plot(t,t**2,'b--',label='^2')
plt.plot(t,t**2.2,'rs',label='^2.2')
plt.plot(t,t**2.5,'g^',label='^2.5')
plt.legend()
plt.show()

```

```python
x1 = [1,2,3,4]
y1 = [1,4,9,16]
x2 = [1,2,3,4]
y2 = [2,4,6,8]
lines = plt.plot(x1,y1,x2,y2)
plt.setp(lines[0],color= 'r',linewidth=2.0,label ='x1y1')
plt.setp(lines[1],color= 'g',linewidth=2.0,label='x2y2')
plt.grid()
plt.legend()
plt.show()
```

```python
# working with multiple figure and axis.
def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)
t1 = np.arange(0.0,5.0,0.1)
t2 = np.arange(0.0,5.0,0.02)

plt.figure(1) # make one figure
plt.subplot(211)
plt.plot(t1,f(t1),'b-')

plt.subplot(212)
plt.plot(t2,np.cos(2*np.pi*t2),'r--')
plt.show()
```
